```python
# DONE, ADDED HEADER FOR REFERENCE

# Dependencies and Setup
import pandas as pd
# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"
# Read Purchasing File and store into Pandas data frame   
purchase_data = pd.read_csv(file_to_load)
# For Reference
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# FINAL

total_people = purchase_data["SN"].drop_duplicates()
total_people2 = total_people.count()
total_dicts = [{"Total Players": total_people2}]
df_people = pd.DataFrame(total_dicts) 
df_people
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Unique_Items = purchase_data["Item Name"].nunique()
Average_Price = purchase_data["Price"].mean()
Number_Purchases = purchase_data["Purchase ID"].count()
Total_Revenue = purchase_data["Price"].sum()

#Create a table as a dataframe 
purchasing_dicts = [{"Number of Unique Items": Unique_Items, "Average Price": Average_Price, "Number of Purchases": Number_Purchases, "Total Revenue": Total_Revenue}]
organized_data_df = pd.DataFrame(purchasing_dicts)
```

```python
# FINAL - Added currency formatting
organized_data_df["Average Price"] = organized_data_df["Average Price"].map("${:.2f}".format)
organized_data_df["Total Revenue"] = organized_data_df["Total Revenue"].map("${:.2f}".format)
organized_data_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results

```python

gender_count = purchase_data.groupby("Gender")
```

```python
gender_total_purchases12 = gender_count.mean()['Price']
```

```python
avg_unique = total_purchase_value / number_unique
```

```python
# FINAL 

bygender_df = pd.DataFrame({"Purchase Count": gender_total_purchases, "Average Purchase Price": gender_total_purchases12, 
              "Total Purchase Value": total_purchase_value, "Avg Total Purchase per Person": avg_unique 
             })

bygender_df["Average Purchase Price"] = bygender_df["Average Purchase Price"].map("${:.2f}".format)
bygender_df["Total Purchase Value"] = bygender_df["Total Purchase Value"].map("${:.2f}".format)
bygender_df["Avg Total Purchase per Person"] = bygender_df["Avg Total Purchase per Person"].map("${:.2f}".format)

bygender_df 
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Use bins  
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ["Under 10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Range1"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
```

```python

age_stuff = purchase_data.groupby("Age Range1")
```

```python
average_purchase_price_age = age_stuff["Price"].mean()
purchase_count_age = age_stuff["Purchase ID"].count()
total_purchase_value_age = age_stuff["Price"].sum()
number_unique_age = age_stuff["SN"].nunique()
average_total_perperson_age = total_purchase_value_age / number_unique_age
```

```python
summary_df2 = pd.DataFrame({'Purchase Count' : purchase_count_age, "Average Purchase Price": 
                            average_purchase_price_age, 
                            "Total Purchase Value": total_purchase_value_age, "Avg Total Purchase per Person": 
                            average_total_perperson_age
                })

summary_df2["Average Purchase Price"] = summary_df2["Average Purchase Price"].map("${:.2f}".format)
summary_df2["Total Purchase Value"] = summary_df2["Total Purchase Value"].map("${:.2f}".format)
summary_df2["Avg Total Purchase per Person"] = summary_df2["Avg Total Purchase per Person"].map("${:.2f}".format)


summary_df2
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#use the (max)

top_spenders = purchase_data.groupby("SN")
```

```python
average_purchase_price_top = top_spenders["Price"].mean()
purchase_count_top = top_spenders["Purchase ID"].count()
total_purchase_value_top = top_spenders["Price"].sum()
```

```python
summary_df3 = pd.DataFrame({"Purchase Count" : purchase_count_top, "Average Purchase Price": 
                            average_purchase_price_top, 
                            "Total Purchase Value": total_purchase_value_top
                })

summary_df4 = summary_df3.sort_values("Total Purchase Value", ascending=False).head(5)

summary_df4["Average Purchase Price"] = summary_df4["Average Purchase Price"].map("${:.2f}".format)
summary_df4["Total Purchase Value"] = summary_df4["Total Purchase Value"].map("${:.2f}".format)


summary_df4
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items = purchase_data.groupby(["Item ID","Item Name"])
```

```python
purchase_price_pop = popular_items["Price"].median()
purchase_count_pop = popular_items["Item ID"].count()
total_purchase_value_pop = popular_items["Price"].sum()
```

```python
summary_df5 = pd.DataFrame({"Purchase Count" : purchase_count_pop, "Item Price": purchase_price_pop, 
                            "Total Purchase Value": total_purchase_value_pop
                })

summary_df6 = summary_df5.sort_values("Purchase Count", ascending=False).head(5)

summary_df6["Item Price"] = summary_df6["Item Price"].map("${:.2f}".format)
summary_df6["Total Purchase Value"] = summary_df6["Total Purchase Value"].map("${:.2f}".format)

summary_df6
```
